```python
# written by Tanner Yilmaz
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')

import_data = pd.read_csv(filepath_or_buffer = "dataexample.csv")

df = pd.DataFrame(data = import_data)
df.head() #shows some of the imported data

x = df["time"]
y = df["temperature"]

plt.plot(x, y, "rx", markersize = 25, mew = 6)
plt.xlabel("time")
plt.ylabel("temperature")
plt.text(7, 12, "Made by Tanner Yilmaz")

#customizing plot size
plt.gcf().set_size_inches(18, 10) 

#customizing font style
font = {'family' : 'monospace',
        'weight' : 'bold',
        'size'   : 20}
plt.rc('font', **font)

plt.savefig("fig_1.png")
```
