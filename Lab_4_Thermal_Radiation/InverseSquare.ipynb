```python
import numpy as np
import matplotlib.pyplot as plt
import math
from linreg import linreg
```

```python
distance = np.array([4,6,8,10,12,15,20,25,30,35,40]) # distance in cm of the radiation sensor from the S-B bulb
radiation_level_mV = np.vstack(([32.4,31.9,33.6],[15.9,15.8,15.5],[8.4,8.3,8.4],[5.3,4.9,5.6],[3.9,3.8,3.8],
                                [2.4,2.2,2.4],[1.3,1.2,1.4],[0.8,0.8,0.8],[0.5,0.5,0.5],[0.3,0.4,0.3],[0.2,0.2,0.2])) # units of mV

radiation_level_mV_err = np.sqrt(0.05**2 + np.std(radiation_level_mV, axis = 1)**2) # summing errors in quadrature
avg_radiation_lvl_mV = np.mean(radiation_level_mV, axis = 1) - (-1) # units of mV. subtracting 1 mV because that is background lvl 
avg_radiation_lvl_err = (3*((1/3)**2)*(radiation_level_mV_err**2))**0.5 # this is error due to lack of precision of voltmeter
```

```python
plt.errorbar(distance, avg_radiation_lvl_mV, yerr = avg_radiation_lvl_err, fmt = "bo", markersize = "5", linewidth=3)
plt.xlabel('Distance (cm)')
plt.ylabel('Radiation level (mV)')
plt.suptitle("Radiation level vs distance")
```

```python
x = np.log(distance)
y = np.log(avg_radiation_lvl_mV)
err_y = np.log(avg_radiation_lvl_err)

coeff = linreg(x, y, err_y, len(x) - 2, 'log(distance)', 'log(radiation level)', "log(radiation level) vs log(distance)")
```
